# Definitions of the variables

PS: Surface Pressure(kPa)
TS: Earth Skin Temperature(C)
PRECTOTCORR: Precipitation Corrrected: (mm/day)
PECTOTCORR_ Sum: Precipitation Corrected Sum(mm)
YEAR: Represents the year in which the climate variable was collected
ANN: The annulled average of the climate variables in question
JAN, FEB, MAR, APR, MAY, JUN, JUL, AUG, SEP, OCT, NOV, DEC: Represents the quantity of the climate variable measured during the month concerned


```python

# Check for missing values in the DataFrame
if data_kaolack.isnull().any().any():
    print("There are missing values in the data")
else:
    print("There are no missing values in the data.")
```

```python
#Extract only the precipitation
rain_kaolack =data_kaolack[data_kaolack['PARAMETER'] == 'PRECTOTCORR_SUM']
```

```python
rain_kaolack.head(2)
```

```python
# Display the dimensions of the dataframe
print("Dimensions : ", rain_kaolack.shape)
```

```python
#Display descriptive statistics for each numeric column
print(rain_kaolack.describe())
```

```python
# Monthly precipitation variation from 1981 to 2020 in the Kaolack region
for i in range(2, len(rain_kaolack.columns)):
    plt.plot(rain_kaolack.iloc[:, 1], rain_kaolack.iloc[:, i], label=rain_kaolack.columns[i])
    
plt.title('"Monthly precipitation variation from 1981 to 2020 in the Kaolack region')
plt.xlabel(rain_kaolack.columns[1])
plt.ylabel('Rain(mm)')
# Move the legend to the right side
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
rain_kaolack_1 = data_kaolack[data_kaolack['PARAMETER'] == 'PRECTOTCORR'][-5:]
# Monthly precipitation variation from 2016 to 2020 in the Kaolack region
for i in range(2, len(rain_kaolack_1.columns)):
    plt.plot(rain_kaolack_1.iloc[:, 1], rain_kaolack_1.iloc[:, i], label=rain_kaolack_1.columns[i])
    
plt.title('"Monthly precipitation variation from 2016 to 2020 in the Kaolack region')
plt.xticks([2016, 2017, 2018, 2019, 2020], [2016, 2017, 2018, 2019, 2020])
plt.xlabel(rain_kaolack_1.columns[1])
plt.ylabel('Rain(mm)')
# Move the legend to the right side
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python

# Calculate the monthly average for each month
monthly_average = rain_kaolack.loc[:, "JAN":"DEC"].mean()

# Draw a bar graph of monthly averages
plt.bar(monthly_average.index, monthly_average)

# Adding axis and title labels
plt.xlabel("Month")
plt.ylabel("Average monthly rainfall (mm)")
plt.title("Monthly average rainfall from 1981 to 2020")

# View diagrame
plt.show()
```
